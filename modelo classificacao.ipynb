```python
#importando bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importando dataset, podemos escolher no site do scikit
from sklearn.datasets import load_iris
data = load_iris()
```

```python
data #aqui é o retorno do load_iris
#data.data #p/ valores de X
#data.target #p/ valores de Y
```

```python
#transformando em dataframe
iris = pd.DataFrame(data.data)
#atraves do print do data na celula acima obtemos os nomes das COLUNAS
#definindo os nomes nas colunas:
iris.columns = data.feature_names
#criando colune
iris['Target'] = data.target
iris.head()
```

```python
#contando cada target, podemos usar para todas as colunas
iris['Target'].value_counts()
#testando para o comprimento das petalas:
iris['petal length (cm)'].value_counts()
```

```python
#filtrando, mostrando soh valores para target = 0
iris[iris['Target'] == 0]
```

```python
#filtro para pegar tudo que é diferente de 2
#para guardar a informacao usamos o sinal de igual no dataframe, igual abaixo:
iris = iris[iris['Target'] != 2]
```

```python
#observando as diversas combinacoes de variaveis explicativas e explicadas
import seaborn as sns
# com hue = 'Target', ele vai colorir os dados, distinguindo os dois tipos de Target
sns.pairplot(iris, hue='Target')
```

```python
#fazendo um scatterplot pelo matplotlib
fig, ax = plt.subplots()

x = iris['petal length (cm)']
y = iris['petal width (cm)']

#lembrando que removemos Target diferente de 2, por isso temos 2 cores diferentes (sobrou Target = 0 e 1)
ax.scatter(x,y,c=iris['Target'])
#plt.show()

#separando por uma reta, criando dois pontos
x_reta = [1,3]
y_reta = [2,0]
#quando X for 1 Y sera 2, quando X for 3 Y sera 0
ax.plot(x_reta,y_reta)
plt.show()
```

```python

fig, ax = plt.subplots()

x = iris['petal length (cm)']
y = iris['petal width (cm)']
ax.scatter(x,y,c=iris['Target'])

#definindo uma reta onde o x vai do 1 ate o 6
x_reta3 = np.arange(1,6)
#com base na equacao da reta que calculamos manualmente antes
y_reta3 = x_reta3*(-1) + 3

ax.plot(x_reta3,y_reta3,c='r')
ax.plot()

#como a equacao da reta deu y = -x + 3 
#para obtermos os pontos em cima da reta, devemos joga o X na formula e verificar qual o Y
#para x = 2, teremos um y = 1, nossa coordenada é (2,1)
ax.scatter(2,1,c='g')
```

```python
#criando uma funcao para classificar o modelo 
def classificacao(x,y):
    y_reta = -x + 3
    y_observacao = y 
    print('Observado {}, Da reta {}'.format(y_observacao,y_reta))

    if y_observacao > y_reta:
        return 1 
    elif y_observacao < y_reta:
        return 0
    else:
        return -1

#testando a funcao
classificacao(5,0.7)
```

```python
#descobrindo a melhor reta com o PERCEPTRON
from sklearn.datasets import load_digits
from sklearn.linear_model import Perceptron


clf = Perceptron(tol=1e-3, random_state=0)
#ferramente de ajuste para o nosso modelo

Xre = iris[['petal length (cm)','petal width (cm)']]
Yre = iris[['Target']]
clf.fit(Xre, Yre)

#imprimindo os pesos em Xre, vao ser dois 
pesos = clf.coef_
#intercepto
intercepto = clf.intercept_

```

```python
# calculando o y do perceptron
# w0 + w1*x + w2*y = 0
# y = (-w0 - w1*x) / w2

# y = (-clf.intercept_-clf.coef_[0][0]*x) / clf.coef_[0][1]
```

```python
#repetindo para comparar o grafico com o novo pos calculos dos pesos
fig, ax = plt.subplots()

x = iris['petal length (cm)']
y = iris['petal width (cm)']
ax.scatter(x,y,c=iris['Target'])

x_reta3 = np.arange(1,6)
#com base na equacao da reta que calculamos antes, reta vermelha
y_reta3 = x_reta3*(-1) + 3
ax.plot(x_reta3,y_reta3,c='r')

x_perceptron = np.arange(1,6)
#com base no calculo dos pesos, reta verde
y_perceptron = (-intercepto-pesos[0][0]*x_perceptron)/pesos[0][1]
ax.plot(x_perceptron,y_perceptron,c='g')
ax.plot()
```

```python
#repetindo para comparar o grafico com o novo pos calculos dos pesos
fig, ax = plt.subplots()

x = iris['petal length (cm)']
y = iris['petal width (cm)']
ax.scatter(x,y,c=iris['Target'])

x_perceptron = np.arange(1,6)
#com base no calculo dos pesos, reta verde
y_perceptron = (-clf.intercept_-clf.coef_[0][0]*x_perceptron)/clf.coef_[0][1]
ax.plot(x_perceptron,y_perceptron,c='g')
ax.plot()

#criando uma funcao para classificar o novo modelo 
def classificacao(y,x_perceptron):
    y_perceptron = (-clf.intercept_-clf.coef_[0][0]*x_perceptron)/clf.coef_[0][1]
    y_observacao = y 
    print('Observado {}, Do perceptron {}'.format(y_observacao,y_perceptron))
    #visualizando o pontinho que colocamos no input
    ax.scatter(x_perceptron,y,c='g')
    #visualizando o ponto para o X da reta
    ax.scatter(x_perceptron,y_perceptron)

    if y_observacao > y_perceptron:
        return 1 
    elif y_observacao < y_perceptron:
        return 0
    else:
        return -1
    
classificacao(1.4,3)
```
